```python
drop_cols = ['No', 'First Name', 'Last Name', 'Start Date']
df.drop(columns=[col for col in drop_cols if col in df.columns], inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders_dict = {}
categorical_columns = ['Gender', 'Department', 'Country', 'Center']
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders_dict[col] = le
```

```python
df.head(10)
```

```python
X=df[['Years', 'Job Rate','Gender', 'Department','Country','Center']]
y=df['Annual Salary']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=23)
X_train
```

```python
len(X_train)

```

```python
import numpy as np
from sklearn.linear_model import LinearRegression,LogisticRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
models={'lr':LinearRegression(), 'log':LogisticRegression(), 'ri':Ridge(), 'la':Lasso(), 'Dtr':DecisionTreeRegressor(),'Rfr':RandomForestRegressor(),'Gbr':GradientBoostingRegressor(),'svr':SVR(),'Knr':KNeighborsRegressor()}
```

```python
results = []
def evaluate_model(name, model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    accuracy = np.mean(np.abs(y_pred - y_test) / y_test < 0.10)* 100
    results.append((name, mae, mse, rmse, accuracy))

for name, model in models.items():
    evaluate_model(name, model)

```

```python
results_df = pd.DataFrame(results, columns=['Model', 'MAE', 'MSE', 'RMSE', 'Accuracy (%)'])
print(results_df.sort_values(by='MAE', ascending=False))
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.barplot(x='Model', y='Accuracy (%)', data=results_df.sort_values(by='Accuracy (%)', ascending=False))
plt.xticks()
plt.title('Model Comparison by Accuracy (%)')
plt.show()

```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Model', y='MAE', data=results_df.sort_values(by='MAE', ascending=False))
plt.xticks()
plt.title('Model Comparison by MAE')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Model', y='MSE', data=results_df.sort_values(by='MSE', ascending=False))
plt.xticks()
plt.title('Model Comparison by MSE')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Model', y='RMSE', data=results_df.sort_values(by='RMSE', ascending=False))
plt.xticks()
plt.title('Model Comparison by RMSE')
plt.show()
```

```python
#choose best model
import joblib
best_model_name = results_df.sort_values(by='Accuracy (%)', ascending=False).iloc[0]['Model']
best_model=models[best_model_name]
joblib.dump(best_model, 'best_salary_model.pkl')
# Save LabelEncoders for later use in Streamlit app
joblib.dump(label_encoders_dict, 'label_encoders.pkl')  # This assumes you stored all encoders in a dictionary
print(f"\n Best model '{best_model_name}' saved as 'best_salary_model.pkl'.")
```

```python
%%writefile app.py
import streamlit as st
import joblib
import pandas as pd
#streamlit app
model = joblib.load('best_salary_model.pkl')
label_encoders = joblib.load('label_encoders.pkl')

st.set_page_config(page_title="Employee Salary Prediction", page_icon="", layout="centered")
st.title("Emloyee Salary Prediction App")
st.markdown("With This App, One Can Predict An Average Salary Of Employees")

# Collect user inputs (raw readable category names)
Gender = st.selectbox("Select Gender", label_encoders['Gender'].classes_)
Department = st.selectbox("Select Department", label_encoders['Department'].classes_)
Country = st.selectbox("Select Country", label_encoders['Country'].classes_)
Center = st.selectbox("Select Center", label_encoders['Center'].classes_)
Years = st.number_input("Years of Experience", min_value=0, max_value=50, value=5)
Job_Rate = st.number_input("Job Rate", min_value=1, max_value=5, value=3)

# Encode user inputs using the saved encoders
Gender_encoded = label_encoders['Gender'].transform([Gender])[0]
Department_encoded = label_encoders['Department'].transform([Department])[0]
Country_encoded = label_encoders['Country'].transform([Country])[0]
Center_encoded = label_encoders['Center'].transform([Center])[0]

# Create DataFrame in model input format
input_df = pd.DataFrame([{
    'Gender': Gender_encoded,
    'Department': Department_encoded,
    'Country': Country_encoded,
    'Center': Center_encoded,
    'Years': Years,
    'Job_Rate': Job_Rate
}])

st.write("### Input Data")
st.write(input_df)
# Predict salary
if st.button("Predict Salary"):
    prediction = model.predict(input_df)
    salary = prediction[0]

    if salary <= 0:
        st.warning(" Predicted Salary: Not Applicable for the given inputs.")
    else:
        st.success(f" Predicted Salary: ${salary:,.2f}")
```
